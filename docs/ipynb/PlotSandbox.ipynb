```python
"""
Adapted from from https://www.youtube.com/watch?v=afITiFR6vfw
First create some data
"""
import random  # already import pyplot above
from matplotlib import style

style.use('fivethirtyeight')

fig = plt.figure()

def create_plot_data(rseed=777):
    random.seed(rseed)
    xs = []
    ys = []
    
    for i in range(10):
        x = i
        y = random.randrange(10)
        
        xs.append(x)
        ys.append(y)
        
    return xs, ys

ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(212)

x,y = create_plot_data(797)
ax1.plot(x,y)

x,y = create_plot_data(327)  # will be different because random values being generated
ax2.plot(x,y)

x,y = create_plot_data(329)
ax3.plot(x,y)

plt.show()
```

```python
# another way: subplot2grid - offers more control
ax1 = plt.subplot2grid((6,1), (0,0), rowspan=1, colspan=1)
ax2 = plt.subplot2grid((6,1), (1,0), rowspan=4, colspan=1)
ax3 = plt.subplot2grid((6,1), (5,0), rowspan=1, colspan=1)

x,y = create_plot_data(227)
ax1.plot(x,y)

x,y = create_plot_data(228)  # will be different because random values being generated
ax2.plot(x,y)

x,y = create_plot_data(229)
ax3.plot(x,y)

plt.show()
```
